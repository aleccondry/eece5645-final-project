```python
#transforming the data in the 1st dataset #diabetic_012 into binary valued, combining prediabetic and diabetic

data['Diabetes_012'].replace(2, 1, inplace=True)
data["Diabetes_012"].hist()
plt.show()
data["Diabetes_012"].value_counts()
data = data.rename(columns={'Diabetes_012':'Diabetes_binary'})
```

```python
data_merged = pd.concat([data,data1], axis=0)
data_merged.info()
```

```python
data_merged["Diabetes_binary"].value_counts()
```

```python
data_merged.hist(figsize = (20,20))
```

```python
columns_to_normalize = ['BMI','MentHlth','PhysHlth']
scaler = StandardScaler()
data_merged[columns_to_normalize] = scaler.fit_transform(data_merged[columns_to_normalize])
data[columns_to_normalize].hist()
```
