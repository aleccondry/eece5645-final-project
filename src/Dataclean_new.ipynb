```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
filepath = '/content/diabetes_012_health_indicators_BRFSS2015.csv'
data = pd.read_csv(filepath)
```

```python
data["Diabetes_012"].hist()
plt.show()
data["Diabetes_012"].value_counts()
```

```python
data1["Diabetes_binary"].hist()
plt.show()
data1["Diabetes_binary"].value_counts()
```

```python
data2["Diabetes_binary"].hist()
plt.show()
data2["Diabetes_binary"].value_counts()
```

```python
#transforming the data in the 1st dataset #diabetic_012 into binary valued, combining prediabetic and diabetic

data['Diabetes_012'].replace(2, 1, inplace=True)
data["Diabetes_012"].hist()
plt.show()
data["Diabetes_012"].value_counts()
data = data.rename(columns={'Diabetes_012':'Diabetes_binary'})
```

```python
#concatenating the first two datasets
data_merged = pd.concat([data,data1], axis=0)
data_merged.info()
```

```python
data_merged["Diabetes_binary"].value_counts()
```

```python
data_merged.hist(figsize = (20,20))
```

```python
#the merged dataset is the training set
columns_to_normalize = ['BMI','MentHlth','PhysHlth']
scaler = StandardScaler()
data_merged[columns_to_normalize] = scaler.fit_transform(data_merged[columns_to_normalize])
data_merged[columns_to_normalize].hist()
```

```python
data_merged.isnull().any()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data_merged.corr(),cmap="YlGnBu",annot=True, annot_kws={"size":8}) 
```

```python
#saving the training dataset into csv
data_merged.to_csv('/content/Training_data.csv', sep=',', index=False, encoding='utf-8')
```
